# Iterating Over Arrays

```python
test = np.random.randint(0, 10, (4,3))
test
```

```python
for row in test:
        print(row)
```

```python
for i in range(len(test)):
    print(test[i])
```

```python
for i, row in enumerate(test):
    print('row', i, 'is', row)
```

```python
test2 = test ** 2
test2
```

```python
for i, j in zip(test, test2):
    print(i, '+', j, '=', i+j)
```
